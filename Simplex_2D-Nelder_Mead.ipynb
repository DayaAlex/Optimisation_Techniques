```python
import numpy as np
import math


iteration = 0
```

```python
#defining the function
def func_val(point):
    himmelblau = (point[0]**2+point[1]-11)**2+(point[0]+point[1]**2-7)**2
    return round(himmelblau,3)
```

```python
#generating simplex from one point
def generate_simplex(point):
    simplex = np.array([point])
    
    point2 = np.copy(point)
    point2[0] = round(point2[0] + delt_1, 3)
    point2[1] = round(point2[1] + delt_2, 3)
    
    point3 = np.copy(point)
    point3[0] = round(point3[0] + delt_2, 3)
    point3[1] = round(point3[1] + delt_2, 3)
    
    simplex = np.append(simplex, [point2], axis = 0)
    simplex = np.append(simplex, [point3], axis = 0)
    
    #simplex = np.append(simplex, [[2.0,0.0]], axis = 0)
    #simplex = np.append(simplex, [[1.0,1.0]], axis = 0)
    
    print('simplex is ', simplex )
    
    idx= ascending(simplex)
    sorted_simplex = [simplex[idx[0]],simplex[idx[1]], simplex[idx[2]]]
    return sorted_simplex
        
```

```python
'''sort the points of simplex according to the function value
first element will have the worst value, next two values correspond to indexes we use to form the line'''

def ascending(simplex):
    values= np.array([])
    
    for (i,point) in enumerate(simplex):
        values= np.append(values,func_val(point))
    print('val:',values)
    print('sorted arguments:',np.sort(values))
    return np.argsort(values)    
    
```

```python
#to calculate centroid of remaining points(leave the worst)
def centroid(simplex):
    x_c =np.array([])
    x_c= 0.5*np.add(simplex[0],simplex[1])
    print('x_c',x_c)
    return x_c
```

```python
'''first we calculate the reflected point of worst value then 
compare its func_val with func_val of the simplex points,
then decide to expland or contract from the reflected point'''

def search_dir(simplex, alph=1.5, bet=0.5):
    global iteration
    iteration+=1
    print('---------------------',iteration,'------------------------')
    
    x_l = simplex[0] # best point having the Least value(minimize)
    x_g = simplex[1] # middle value point
    x_h = simplex[2] # worst point having the largest value(minimize)
    x_c = centroid(simplex)
    
    x_r = 2*x_c - x_h # calculing the reflected point, r = 2
    print('x_r:',x_r)
    print('func val of x_r:',func_val(x_r))
    print('func val of x_l:',func_val(x_l))
    print('func val of x_g:',func_val(x_g))
    print('func val of x_h:',func_val(x_h))
    #expand in the direction of reflected point if less than best point, farther in that direction
    if (func_val(x_r)<func_val(x_l)):
        x_new = x_c + alph*(x_c-x_h)
        
    #contract inward the centroid point if func val of reflected point is greater than the worst point
    elif (func_val(x_r)>func_val(x_h)):  
        x_new = x_c - bet*(x_c - x_h)
    
    #expand little outward the centroid point, less than the reflected point, termed as contracting outward
    else:  
        x_new = x_c + bet*(x_c - x_h)
    x_h = x_new
    print('func val of x_new:',func_val(x_new))
    
    new_simplex=[x_l,x_g,x_h]
    idx= ascending(new_simplex)
    sorted_simplex = [new_simplex[idx[0]],new_simplex[idx[1]], new_simplex[idx[2]]]
    print('new simplex is:', sorted_simplex, 'new centroid',x_c)
    return sorted_simplex, x_c
    
```

```python
def optimize(simplex,cent,tolerance=10e-2):
    ss_sum= 0
    ss_sum = np.sum((func_val(element) - func_val(cent))**2 for element in simplex)
    sq_sum=round(math.sqrt(ss_sum/3),3)
    
    if sq_sum<=tolerance:
        print('sq_sum is :',sq_sum)
        print('tolerance is :',tolerance)
        
        print('optima is :', func_val(simplex[0]),'at x = ',simplex[0])
    else:
        simplex, cent = search_dir(simplex)
        optimize(simplex,cent)
        
```

```python
point = [0.0,0.0]
tolerance = 10e-2
alph = 1.5
bet = 0.5

#assuming scale factor =2 
delt_1 = (math.sqrt(3)+1)/math.sqrt(2) #delt_1 = {((n-1)^0.5+n-1)/(n(2)^0.5)}lambda
delt_2 = (math.sqrt(3)-1)/math.sqrt(2) #delt_2 = {((n-1)^0.5-1)/(n(2)^0.5)}lambda

simp = generate_simplex(point)

simpl, cent = search_dir(simp,alph,bet)
optimize(simpl,cent,tolerance)
        
     
```
