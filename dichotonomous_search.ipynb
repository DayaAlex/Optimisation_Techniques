# Neccessary Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Defining the objective and visualising it

```python
iteration =0
function_call=0

def func(x):
  global function_call
  function_call+=1
  return 5*x**2 - 3*x + 2

#visualising the function for the bounds [-4,4]

x = np.linspace(-5,5)
y = func(x)

plt.plot(x,y,'blue')
plt.axvline(x=4,color='red',linestyle=':',label ='b=4')
plt.axvline(x=-4,color='red',linestyle=':',label='a=-4')
plt.title('5x^2-3x+2 for the domain [-4,4]')
plt.legend()

```

# Defining dichotonomous function

```python
def min_dichot(low_bound,up_bound,epsilon=10e-4):
  global iteration, function_call
  interval = (up_bound - low_bound)*0.5
  if interval< epsilon:#checking if required accuracy is achieved.
    return (low_bound+up_bound)*0.5,(low_bound,up_bound)
  else:
    mid = (low_bound+up_bound)*0.5
    iteration+=1
    c = mid -(epsilon*0.5)
    d = mid +(epsilon*0.5)
    #eliminating close to 50% of current search interval depending on function values
    if func(c)<func(d):
      up_bound = d
      return min_dichot(low_bound,up_bound,epsilon)
    else:
      low_bound = c
      return min_dichot(low_bound,up_bound,epsilon)

```

# Calling the function

```python
epsilon=10e-6

min_val,(low_bound,up_bound) = min_dichot(-4,4,epsilon)
print('minimum of the function is : ',min_val)
print('Number of iteration:',iteration)
print('Number of function calls:',function_call)
```

# Plotting the optimum

```python
plt.plot(x,y,'blue')
plt.axvline(x=4,color='red',linestyle=':',label ='b=4')
plt.axvline(x=-4,color='red',linestyle=':',label='a=-4')
plt.axvline(x=low_bound,color='green', linestyle='-.', label='c = lower bound')
plt.axvline(x=up_bound,color='yellow', linestyle='-.', label='d = upper bound')
plt.title('5x^2-3x+2 for the domain [-4,4]')
plt.legend()
```
